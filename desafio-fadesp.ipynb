    - Realize o processo de extração do dataset diretamente a partir do repositório do github utilizando Python

```python
import pandas as pd
import requests

# URL do arquivo CSV no GitHub
url = 'https://raw.githubusercontent.com/raphagama/desafio-fadesp-processo-seletivo/master/data/dataset_desafio_fadesp.csv'

# Fazendo o download do arquivo
response = requests.get(url)

# Verificando se o download foi bem-sucedido
if response.status_code == 200:
    # Carregando os dados em um DataFrame
    dados = pd.read_csv(url, encoding='latin1')
    # Agora você pode prosseguir com a análise dos dados
else:
    print("Falha ao baixar o arquivo.")

dados.head()
```

- **OBJETIVO**
    2. Realize um processo limpeza nos dados removendo ou substituindo os valores nulos quando necessário, e eliminando dados duplicados e inconsistentes caso exista algum.

```python
# Remoção de valores nulos
dados_limpos = dados.dropna()

# Remoção de dados duplicados
dados_limpos = dados_limpos.drop_duplicates()

#Mostrando todas as colunas para tratamento no próximo passo
pd.set_option('display.max_columns', None)

print(dados_limpos)
```
